```python
import pandas as pd
import pysam
from tqdm import tqdm

# Load the .cram file
cram_file = pysam.AlignmentFile("data/ADGeneData/AMPAD_genomicdata/R1913204-DLPFC.final.cram", "rc")

# Set the gene details manually
gene_name = 'APOE'
chrom = 'chr19'
start = 44905791
end = 44909393

# Initialize a list to hold the read data for this gene
gene_data = []

# Fetch alignments in the region of interest
for read in tqdm(cram_file.fetch(chrom, start, end), desc="Processing reads"):
    # Pull out the relevant information from the read
    data = {
        'query_name': read.query_name,
        'sequence': read.query_sequence,
        'mapping_quality': read.mapping_quality,
        'reference_start': read.reference_start,
        'cigar_string': read.cigarstring,
    }
    gene_data.append(data)

# Convert the gene data to a DataFrame and save to a CSV file
df = pd.DataFrame(gene_data)

# Close the .cram file when you're done
cram_file.close()
```
