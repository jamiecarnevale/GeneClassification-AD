```python
import pandas as pd
```

```python
# Read the csv files back into pandas dataframes
df_ROSMAP_female = pd.read_csv('filtered_female_ROSMAP.csv')
df_ROSMAP_male = pd.read_csv('filtered_male_ROSMAP.csv')

df_MAYO_female = pd.read_csv('filtered_female_MAYO.csv')
df_MAYO_male = pd.read_csv('filtered_male_MAYO.csv')

df_MSBB_female = pd.read_csv('filtered_female_MSBB.csv')
df_MSBB_male = pd.read_csv('filtered_male_MSBB.csv')

```

```python
# Find common genes in all studies for female
common_genes_female = set(df_ROSMAP_female['hgnc_symbol']).intersection(
    set(df_MAYO_female['hgnc_symbol']), set(df_MSBB_female['hgnc_symbol']))

# Find common genes in all studies for male
common_genes_male = set(df_ROSMAP_male['hgnc_symbol']).intersection(
    set(df_MAYO_male['hgnc_symbol']), set(df_MSBB_male['hgnc_symbol']))

# Convert to DataFrames if needed
df_common_genes_female = pd.DataFrame(list(common_genes_female), columns=['hgnc_symbol'])
df_common_genes_male = pd.DataFrame(list(common_genes_male), columns=['hgnc_symbol'])
```

```python
# Define function to get common genes
def get_common_genes(df1, df2):
    return set(df1['hgnc_symbol']).intersection(set(df2['hgnc_symbol']))

# Get common genes
common_genes_ROSMAP_MAYO = get_common_genes(df_ROSMAP_female, df_MAYO_female)
common_genes_ROSMAP_MSBB = get_common_genes(df_ROSMAP_female, df_MSBB_female)
common_genes_MAYO_MSBB = get_common_genes(df_MAYO_female, df_MSBB_female)

# Store in dictionary
common_genes_dict = {
    'ROSMAP and MAYO': common_genes_ROSMAP_MAYO,
    'ROSMAP and MSBB': common_genes_ROSMAP_MSBB,
    'MAYO and MSBB': common_genes_MAYO_MSBB
}

# Convert to DataFrames if needed
df_common_genes_ROSMAP_MAYO = pd.DataFrame(list(common_genes_ROSMAP_MAYO), columns=['hgnc_symbol'])
df_common_genes_ROSMAP_MSBB = pd.DataFrame(list(common_genes_ROSMAP_MSBB), columns=['hgnc_symbol'])
df_common_genes_MAYO_MSBB = pd.DataFrame(list(common_genes_MAYO_MSBB), columns=['hgnc_symbol'])

df_common_genes_ROSMAP_MAYO['Studies'] = 'ROSMAP and MAYO'
df_common_genes_ROSMAP_MSBB['Studies'] = 'ROSMAP and MSBB'
df_common_genes_MAYO_MSBB['Studies'] = 'MAYO and MSBB'

# Combine all into one DataFrame
df_all_common_genes_female = pd.concat([df_common_genes_ROSMAP_MAYO, df_common_genes_ROSMAP_MSBB, df_common_genes_MAYO_MSBB])
```

```python
# Define function to get common genes
def get_common_genes(df1, df2):
    return set(df1['hgnc_symbol']).intersection(set(df2['hgnc_symbol']))

# Get common genes
common_genes_ROSMAP_MAYO = get_common_genes(df_ROSMAP_male, df_MAYO_male)
common_genes_ROSMAP_MSBB = get_common_genes(df_ROSMAP_male, df_MSBB_male)
common_genes_MAYO_MSBB = get_common_genes(df_MAYO_male, df_MSBB_male)

# Store in dictionary
common_genes_dict = {
    'ROSMAP and MAYO': common_genes_ROSMAP_MAYO,
    'ROSMAP and MSBB': common_genes_ROSMAP_MSBB,
    'MAYO and MSBB': common_genes_MAYO_MSBB
}

# Convert to DataFrames if needed
df_common_genes_ROSMAP_MAYO = pd.DataFrame(list(common_genes_ROSMAP_MAYO), columns=['hgnc_symbol'])
df_common_genes_ROSMAP_MSBB = pd.DataFrame(list(common_genes_ROSMAP_MSBB), columns=['hgnc_symbol'])
df_common_genes_MAYO_MSBB = pd.DataFrame(list(common_genes_MAYO_MSBB), columns=['hgnc_symbol'])

df_common_genes_ROSMAP_MAYO['Studies'] = 'ROSMAP and MAYO'
df_common_genes_ROSMAP_MSBB['Studies'] = 'ROSMAP and MSBB'
df_common_genes_MAYO_MSBB['Studies'] = 'MAYO and MSBB'

# Combine all into one DataFrame
df_all_common_genes_male = pd.concat([df_common_genes_ROSMAP_MAYO, df_common_genes_ROSMAP_MSBB, df_common_genes_MAYO_MSBB])
```

```python
# Convert the 'hgnc_symbol' columns to sets for easy comparison
set_male = set(df_all_common_genes_male['hgnc_symbol'])
set_female = set(df_all_common_genes_female['hgnc_symbol'])

# Find the intersection (common genes)
common_genes = set_male.intersection(set_female)

# Find the difference (sex-specific genes)
male_specific_genes = set_male.difference(set_female)
female_specific_genes = set_female.difference(set_male)

# Convert back to dataframes for further analysis or saving to CSV
df_common_genes = pd.DataFrame(list(common_genes), columns=['hgnc_symbol'])
df_male_specific_genes = pd.DataFrame(list(male_specific_genes), columns=['hgnc_symbol'])
df_female_specific_genes = pd.DataFrame(list(female_specific_genes), columns=['hgnc_symbol'])
```
